```python
def open_docx(path):
  import docx
  doc = docx.Document(path)
  list_of_content = []
  for para in doc.paragraphs:
      list_of_content.append(para.text)

  return list_of_content
```

```python
row = []
for i in range(len([x[0] for x in os.walk(ROOT_DIRECTORY)])):
  filelist = os.listdir([x[0] for x in os.walk(ROOT_DIRECTORY)][i])
  for f in filelist:
    if f.endswith(".docx"):
      doc = open_docx([x[0] for x in os.walk(ROOT_DIRECTORY)][i] + '/' + f)
      text = " ".join(doc)
      text = text.lower()
      text = re.sub('[^.آ-یa-z]+', ' ', text)
      text = re.sub("\u200c", " ", text)
      text = re.sub("\s+", " ", text)
      text
      row.append(
          {
              "name": str([x[0] for x in os.walk(ROOT_DIRECTORY)][i]).split('/')[-1],
              "content": text
          }
      )

job_positions = pd.DataFrame(row)
job_positions
```

```python
skills = []
jobs = []

for row in job_positions.itertuples():
  tmp_resume = row.content
  tmp_resume = remove_stopwords(tmp_resume)
  tmp_resume = re.sub("[.]", "", tmp_resume)
  tmp_resume = re.sub("\s+", " ", tmp_resume)

  skills.append(
     "-".join(list(set([x['word'] for x in pipe(tmp_resume) if x['entity_group'] == 'skill'])))
  )
  jobs.append(
      "-".join(list(set([x['word'] for x in pipe(tmp_resume) if x['entity_group'] == 'job_title'])))
  )
```

```python
job_positions['skills'] = skills
job_positions['jobs'] = jobs
job_positions
```

```python
string_feature = []

for index, row in job_positions.iterrows():


    combined = ""

    if not row["skills"] == "":
        combined += f"""The required skills for an applicant are {" ".join(sorted((row['skills']).split("-")))}. """


    if not row["jobs"] == "":
      combined += f"""job title is {" ".join(sorted((row['jobs']).split("-")))}"""

    string_feature.append(combined)
```

```python
job_positions['string_feature'] = string_feature
```
