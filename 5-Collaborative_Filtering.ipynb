## Create user-item matrix

```python
items = []
users = []
ratings = []
for x in classification_df.itertuples():
  for y in job_positions.itertuples():
    items.append(y.name)
    users.append(x.id)
    score = 0
    if x.category == y.name:
      score = 1 + calculate_cosine_sim(sentence_embedding(x.string_feature),sentence_embedding(y.string_feature))
    ratings.append(score)
```

```python
from surprise import Dataset
from surprise import Reader


ratings_dict = {
    "item": items,
    "user": users,
    "rating": ratings,
}

df = pd.DataFrame(ratings_dict)
reader = Reader(rating_scale=(1, 5))


data = Dataset.load_from_df(df[["user", "item", "rating"]], reader)
```
